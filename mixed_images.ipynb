```python
import os
import numpy as np
from PIL import Image
import shutil

# === Input Paths ===
imageDir = r"c:\Users\wasif\Desktop\SemanticSegmentationData\images"
maskDir = r"c:\Users\wasif\Desktop\SemanticSegmentationData\masks"

# === Output Paths for mixed ===
mixedImgDir = r"c:\Users\wasif\Desktop\SemanticSegmentationData\filtered_mixed\images"
mixedMaskDir = r"c:\Users\wasif\Desktop\SemanticSegmentationData\filtered_mixed\masks"
os.makedirs(mixedImgDir, exist_ok=True)
os.makedirs(mixedMaskDir, exist_ok=True)

# === Filter logic ===
for fname in sorted(os.listdir(maskDir)):
    if not fname.endswith(".png"):
        continue

    maskPath = os.path.join(maskDir, fname)
    mask = np.array(Image.open(maskPath).convert("L"))

    uniqueVals = np.unique(mask)

    if 0 in uniqueVals and 255 in uniqueVals:
        # Mixed image — copy image and mask
        imgName = fname.replace(".png", ".jp2")
        imgPath = os.path.join(imageDir, imgName)

        if os.path.exists(imgPath):
            shutil.copy(imgPath, os.path.join(mixedImgDir, imgName))
            shutil.copy(maskPath, os.path.join(mixedMaskDir, fname))
            print(f"✅ Copied mixed: {imgName}")
        else:
            print(f"⚠️ Missing image for mask: {fname}")

print("\n🎉 Done filtering mixed image-mask pairs.")
```
